Problem Statement 1:Smart Parking Lot Management System


```python
def smart_parking(capacity, vehicle_logs):
    current_parked = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_parked += 1
        elif log == "OUT" and current_parked > 0:
            current_parked -= 1

        # Track peak usage
        peak_usage = max(peak_usage, current_parked)

    print("Currently Parked Vehicles:", current_parked)

    if current_parked > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)


# Sample Input
smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

Problem Statement 2:Online Food Delivery Time Estimator


```python
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5  # Assume 5 minutes per km

    # Traffic adjustment
    if traffic.lower() == "high":
        base_time += 15
    elif traffic.lower() == "medium":
        base_time += 8

    # Weather adjustment
    if weather.lower() == "rainy":
        base_time += 10
    elif weather.lower() == "stormy":
        base_time += 20

    print("Estimated Delivery Time:", base_time, "minutes")


# Sample Input
estimate_delivery_time(8, "High", "Rainy")
```

Problem Statement 3:Movie Theatre Seat Occupancy Analyzer


```python
def seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy_percentage = (booked_count / total_seats) * 100

    print("Occupancy:", int(occupancy_percentage), "%")

    if occupancy_percentage == 100:
        print("Show Status: Housefull")
    elif occupancy_percentage >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")

    if occupancy_percentage > 90:
        print("Suggestion: Consider opening additional shows")


# Sample Input (150 booked seats)
seat_occupancy(200, [1] * 150)
```

Problem Statement 4:Cloud Server Load Classification System

```python
def classify_server_load(cpu_readings):
    average_load = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", int(average_load), "%")

    if average_load < 50:
        print("Server Status: Normal")
    elif 50 <= average_load <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
classify_server_load([45, 60, 70, 85, 90])
```

Problem Statement 5:Smart Classroom Resource Usage Monitor


```python
def classroom_monitor(resource_usage):
    overused = []

    for resource, hours in resource_usage.items():
        if hours > 8:  # Assume 8 hours is limit
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
usage_data = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_monitor(usage_data)
```

Problem Statement 6: Online Event Registration Capacity Controller


```python
def event_registration(capacity, registrations):
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
