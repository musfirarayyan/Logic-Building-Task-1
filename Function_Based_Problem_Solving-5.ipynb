Problem 1: Social Media – Post Engagement Analyzer


```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    return total_likes, status


# Example
likes = [200, 300, 150, 400]
total, result = analyze_post_engagement(likes)
print("Total Likes:", total)
print("Post Status:", result)
```

Problem 3: Agriculture – Rainfall Adequacy Checker


```python
def rainfall_checker(rainfall_list, required_level):
    total = 0

    for rain in rainfall_list:
        total += rain

    average = total / len(rainfall_list)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    return average, status


# Example
rain_data = [60, 75, 80, 70, 75]
avg, status = rainfall_checker(rain_data, 70)
print("Average Rainfall:", avg)
print("Rainfall Status:", status)
```

Problem 4: Social Media – Duplicate Account Detection


```python
def detect_duplicate_usernames(usernames):
    if len(usernames) != len(set(usernames)):
        return "Yes"
    else:
        return "No"


# Example
users = ["john", "mary", "john", "alex"]
result = detect_duplicate_usernames(users)
print("Duplicate Accounts Found:", result)
```

Problem 6: Agriculture – Premium Crop Price Filter


```python
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    return premium


# Example
crop_prices = [1500, 2500, 1800, 3200]
premium_list = filter_premium_crops(crop_prices)
print("Premium Crops:", premium_list)
```

Problem 7: System Monitoring – Application Health Checker


```python
def check_application_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    return errors, status


# Example
errors, status = check_application_health(7)
print("Error Count:", errors)
print("System Status:", status)
```

Problem 9: E-Learning – Student Attendance Eligibility System


```python
def attendance_checker(attendance_list):
    total_classes = len(attendance_list)
    attended = 0

    for status in attendance_list:
        if status == 1:
            attended += 1

    percentage = (attended / total_classes) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    return percentage, eligibility


# Example
attendance = [1, 1, 0, 1, 1]  # 1 = Present, 0 = Absent
percent, status = attendance_checker(attendance)
print("Attendance Percentage:", percent)
print("Exam Eligibility:", status)
```

Problem 10: Smart Electricity Bill Analyzer


```python
def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Example
bill, status = calculate_electricity_bill(250)
print("Total Bill:", bill)
print("Usage Category:", status)
```
