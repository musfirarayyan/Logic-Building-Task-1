**Problem 1: Mobile Recharge Validation System**


 Logic Explanation:



Create a function validate_recharge(amount)

Check:

Amount ≥ 50

Amount must be in valid plans list

Return True if valid, else False

Use while loop to allow retry until valid recharge

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount >= 50 and amount in valid_plans:
        return True
    else:
        return False


while True:
    amount = int(input("Enter recharge amount: ₹"))

    if validate_recharge(amount):
        print("Recharge Successful!")
        break
    else:
        print("Invalid plan. Please try again.")
```

**Problem 2: Inventory Reorder Alert System**

Logic Explanation:



Create function check_inventory(stock_dict)

Loop through dictionary using for

If stock < 15 → print Reorder Alert

Else → print Stock OK

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")


inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 8,
    "Monitor": 20
}

check_inventory(inventory)
```

**Problem 3: Student Result Processing System**


 Logic Explanation:



Create function calculate_result(marks_list)

Use loop to calculate total

Find average

If average ≥ 50 → Pass else Fail

Return result

```python
def calculate_result(marks):
    total = 0

    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


student_marks = [60, 45, 70, 55]
result = calculate_result(student_marks)

print("Result:", result)
```

**Problem 4: Cab Fare Estimator with Retry Option**


Logic Explanation:



Base fare = 50

Fare = 50 + (12 × distance)

If peak hour → add 25% extra

Use while loop for retry

```python
def calculate_fare(distance, peak):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak:
        fare += fare * 0.25

    return fare


while True:
    distance = float(input("Enter distance in km: "))
    peak_input = input("Is it peak hour? (yes/no): ").lower()

    peak = True if peak_input == "yes" else False

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        break
```

**Problem 5: Employee Attendance Eligibility Checker**


 Logic Explanation:



Function check_attendance(attendance_list)

Count "P" using loop

Calculate percentage

Return Eligible if ≥ 75%

```python
def check_attendance(attendance):
    present_days = 0

    for day in attendance:
        if day == "P":
            present_days += 1

    percentage = (present_days / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]
status = check_attendance(attendance_list)

print("Attendance Status:", status)
```

**Problem 6: Password Strength Checker**


 Logic Explanation:



Minimum length check

Loop through characters:

Check if digit exists

Check if special character exists (@#$)

If all conditions satisfied → Strong password

```python
def check_password(password):
    if len(password) < 8:
        return "Weak Password"

    has_digit = False
    has_special = False

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


pwd = input("Enter password: ")
print(check_password(pwd))
```
