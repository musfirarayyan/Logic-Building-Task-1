Problem 1: Unique Words in a Sentence Logic:

---



Take input sentence

Split into words using split()

Convert list into set() to remove duplicates

Count using len():

```python
# Problem 1: Unique Words in a Sentence

sentence = input("Enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert to set to remove duplicates
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

 Problem 2: Highest Salary from Employee Data Logic:

Store data in dictionary

Use max() with key=employees.get

```python
# Problem 2: Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)

print("Highest Salary:", highest_employee, "-", employees[highest_employee])
```

Problem 3: Find Maximum and Minimum Values Logic:

Initialize max and min with first element

Compare using loop



```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("List:", numbers)
print("Max:", max_value)
print("Min:", min_value)
```

Problem 4: Count Products Above Price Threshold Logic:

Loop through list

Increase counter if price > 1000

```python
# Problem 4: Count Products Above a Price Threshold

prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Products above 1000:", count)
```

Problem 5: Calculate Attendance Percentage Logic:

Count "P"

Divide by total length

Multiply by 100

```python
# Problem 5: Calculate Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

percentage = (present_count / total_days) * 100

print("Attendance Percentage:", percentage)
```

Problem 7: Count Character Frequency Logic:

Create empty dictionary

Increase count for each character

```python
# Problem 7: Count Character Frequency

string = "python"

char_count = {}

for char in string:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1

print(char_count)
```
