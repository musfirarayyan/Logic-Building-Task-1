```python
username = "admin"
password = "1234"

input_username = input("Enter username: ")
input_password = input("Enter password: ")

if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print("Total Students:", len(marks))
print("Pass Students:", pass_count)
print("Fail Students:", fail_count)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("Cleaned Names:", cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)
    if length > 10:
        print("Flag: Message longer than 10 characters")

    print()
```

```python
# Given list of log entries
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Initialize counter for ERROR logs
error_count = 0

# Loop through each log entry
for log in logs:
    # Check if the log entry is "ERROR"
    if log == "ERROR":
        error_count += 1

# Print the total number of ERROR entries
print("Total ERROR entries:", error_count)
```
