**Problem 1: Employee Performance Bonus Eligibility**


Logic Explanation:



Use max() to find the highest score.

Loop through dictionary items.

If score equals highest score → store employee name.

```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Step 1: Find highest score
highest_score = max(employees.values())

# Step 2: Find employees with highest score
top_performers = []

for name, score in employees.items():
    if score == highest_score:
        top_performers.append(name)

# Step 3: Display result
print("Top Performers Eligible for Bonus:",
      ", ".join(top_performers),
      f"(Score: {highest_score})")
```

**Problem 2: Search Query Keyword Analysis**


Logic Explanation:



Convert sentence to lowercase → lower()

Remove punctuation → translate()

Split into words → split()

Use dictionary to count frequency.

Print words with frequency > 1.

```python
import string

query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

words = query.split()

frequency = {}

# Count word frequency
for word in words:
    frequency[word] = frequency.get(word, 0) + 1

# Filter words searched more than once
result = {}

for word, count in frequency.items():
    if count > 1:
        result[word] = count

print(result)
```

**Problem 3: Sensor Data Validation**

 Logic Explanation:



Use enumerate() to get index and value.

Check if number is even → value % 2 == 0

Store valid readings as (index, value) tuple.

Ignore odd values.

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for index, value in enumerate(sensor_readings):
    if value % 2 == 0:
        valid_readings.append((index, value))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

**Problem 4: Email Domain Usage Analysis**


 Logic Explanation:



Split email using "@".

Extract domain name.

Count frequency using dictionary.

Calculate percentage:
percentage = (count / total_users) * 100

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

domain_count = {}

# Count domains
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

total_users = len(emails)

# Display percentage
for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {percentage:.0f}%")
```

**Problem 5: Sales Spike Detection**


Logic Explanation:



Calculate average using:
average = sum(sales) / len(sales)

Find threshold:
threshold = average * 1.30

Loop through list with day number.

Print days where sales > threshold.

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

average_sales = sum(sales) / len(sales)
threshold = average_sales * 1.30

print("Sales Spikes Detected:")

for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

**Problem 6: Duplicate User ID Detection**

 Logic Explanation:



Use dictionary to count frequency.

Loop through user IDs.

Increase count for each occurrence.

Print IDs where count > 1.

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

id_count = {}

# Count occurrences
for user in user_ids:
    id_count[user] = id_count.get(user, 0) + 1

# Display duplicates
for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
